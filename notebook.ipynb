```python
import gradio as gr
import tensorflow as tf
import numpy as np

# Load the saved model
model = tf.keras.models.load_model('diabetes_model.h5')

def predict_diabetes(pregnancies, glucose, blood_pressure, skin_thickness, 
                    insulin, bmi, diabetes_pedigree, age):
    # Create input array without scaling
    input_data = np.array([[pregnancies, glucose, blood_pressure, skin_thickness,
                           insulin, bmi, diabetes_pedigree, age]])
    
    # Make prediction directly
    prediction = model.predict(input_data)[0][0]
    
    # Convert to percentage and determine result
    probability = prediction * 100
    result = "High Risk" if prediction > 0.5 else "Low Risk"
    
    return f"{result} (Probability: {probability:.2f}%)"

# Create Gradio interface
interface = gr.Interface(
    fn=predict_diabetes,
    inputs=[
        gr.Number(label="Pregnancies", minimum=0, maximum=20, value=1),
        gr.Number(label="Glucose Level", minimum=0, maximum=300, value=120),
        gr.Number(label="Blood Pressure", minimum=0, maximum=200, value=80),
        gr.Number(label="Skin Thickness", minimum=0, maximum=100, value=25),
        gr.Number(label="Insulin", minimum=0, maximum=1000, value=100),
        gr.Number(label="BMI", minimum=0, maximum=70, value=25),
        gr.Number(label="Diabetes Pedigree Function", minimum=0, maximum=3, value=0.5, step=0.01),
        gr.Number(label="Age", minimum=18, maximum=120, value=30)
    ],
    outputs=gr.Text(label="Diabetes Risk Prediction"),
    title="Diabetes Risk Predictor",
    description="Enter patient information to predict diabetes risk using our trained neural network model.",
    examples=[
        [6, 148, 72, 35, 0, 33.6, 0.627, 50],
        [1, 85, 66, 29, 0, 26.6, 0.351, 31],
        [0, 137, 40, 35, 168, 43.1, 2.288, 33]
    ]
)

# Launch the interface
interface.launch(share=True)

```
